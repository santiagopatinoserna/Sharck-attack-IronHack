# ANALISIS DE ATAQUE DE TIBURONES PARA EMPRESA DE SEGUROS

Este proyecto, realizado en el Ironhack Bootcamp, tiene como objetivo ayudar a una empresa de seguros a optimizar sus ofertas para actividades acuáticas en zonas con alta incidencia de ataques de tiburones. A través del análisis de datos, se identificaron patrones clave sobre ubicaciones, estacionalidad y gravedad de los ataques para desarrollar estrategias efectivas.

Objetivos Principales:
Identificar los países con mayor número de ataques.
Determinar las épocas del año con mayor probabilidad de incidencias.
Evaluar la gravedad de los ataques (fatales o no fatales) para definir pólizas adecuadas.
Herramientas y Técnicas:
Python, Pandas, Matplotlib y Seaborn: Para el análisis, limpieza y visualización de datos.
Fuzzywuzzy y Regex: Para estandarizar datos textuales y fechas.
GitHub y Visual Studio Code: Para colaboración y desarrollo.
Proceso de Limpieza de Datos:
Eliminación de columnas irrelevantes.
Estandarización de nombres y formato de fechas.
Manejo de valores nulos y separación de fechas en meses y años.
Hallazgos Clave:
Países Prioritarios: Estados Unidos, Sudáfrica y Australia son las ubicaciones con mayor número de ataques.
Estacionalidad: Los ataques se concentran en los meses de verano (junio a septiembre en el hemisferio norte).
Gravedad: La mayoría de los ataques no son fatales, lo que sugiere la viabilidad de pólizas de bajo riesgo.
Recomendaciones:
Focalizar campañas en los países con más ataques.
Ofrecer pólizas para accidentes no fatales en actividades acuáticas.
Incrementar esfuerzos publicitarios en primavera y principios de verano.
Este análisis permite a la empresa aseguradora dirigir sus recursos de manera estratégica, enfocándose en las áreas y periodos de mayor riesgo.
